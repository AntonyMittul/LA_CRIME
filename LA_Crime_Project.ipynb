```python
# prompt: identify the top three crimes with the highest average victim age

average_age_by_crime = crime.groupby('Crm Cd Desc')['Vict Age'].mean().reset_index()

# Sort the crimes by average victim age in descending order and get the top 3
top_3_crimes_by_age = average_age_by_crime.sort_values(by='Vict Age', ascending=False).head(3)

print("Top 3 crimes with the highest average victim age:")
top_3_crimes_by_age
```

```python
# prompt: find the areas with the highest number of crimes occuring during night time

import pandas as pd
import matplotlib.pyplot as plt
# Define night time as between 8 PM (20:00) and 5 AM (5:00)
crime['TIME OCC'] = crime['TIME OCC'].astype(str).str.zfill(4)
crime['hour'] = pd.to_datetime(crime['TIME OCC'], format='%H%M').dt.hour

night_crimes = crime[((crime['hour'] >= 20) | (crime['hour'] < 5))]

# Find the areas with the highest number of night crimes
areas_with_most_night_crimes = night_crimes['AREA NAME'].value_counts().head(10)

print("Areas with the highest number of crimes occurring during night time:")
print(areas_with_most_night_crimes)

# Optional: Visualize the areas with most night crimes
plt.figure(figsize=(10, 6))
areas_with_most_night_crimes.plot(kind='bar')
plt.title('Top 10 Areas with Most Night Crimes')
plt.xlabel('Area Name')
plt.ylabel('Number of Night Crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: find the crime counts occuring every hour

import matplotlib.pyplot as plt
crime_count_per_hour = crime['hour'].value_counts().sort_index()

print("Number of crimes occurring every hour:")
print(crime_count_per_hour)

# Optional: Visualize the crime counts per hour
plt.figure(figsize=(10, 6))
crime_count_per_hour.plot(kind='bar', color='teal')
plt.title('Number of Crimes per Hour')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
